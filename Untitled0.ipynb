```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
valores = {
    'lista1': [5,7,9,10] # valores 5, 7-> 9 -> 1 -> 100
}
valores = pd.DataFrame(valores)
```

```python
media = valores.lista1.mean()

plt.hist(valores.lista1);
plt.axvline(x=media, color='r', linestyle='-')
plt.show()

print('m√©dia: '+ str(media) )
```

```python
valores = {
    'lista2': [5,7,9,100] # valores 5, 7-> 9 -> 1 -> 100
}

valores = pd.DataFrame(valores)
```

```python
media = valores.lista2.mean()
mediana = valores.lista2.median()
moda = valores.lista2.mode()
desvio_padrao = valores.lista2.std()

plt.hist(valores.lista2);
plt.axvline(x=media, color='r', linestyle='-')
plt.axvline(x=mediana, color='b', linestyle='-')
plt.axvline(x=moda[0], color='g', linestyle='-')
plt.show()

print('m√©dia: '+ str(media) )
print('mediana: '+ str(mediana) )
print('moda: '+ str(moda) )
print('desvio padr√£o: '+ str(desvio_padrao) )
```
